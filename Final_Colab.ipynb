````python
# Handling Outliers & Outlier treatments
import pandas as pd
import numpy as np

# Load your dataset into DataFrame df
df = pd.read_csv('/content/Housing.csv')

def detect_outliers_iqr(data, k=1.5):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - k * iqr
    upper_bound = q3 + k * iqr
    return (data < lower_bound) | (data > upper_bound)

# Remove outliers using IQR method
def remove_outliers_iqr(data, k=1.5):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - k * iqr
    upper_bound = q3 + k * iqr
    return data[(data >= lower_bound) & (data <= upper_bound)]

# Apply outlier handling techniques to numerical features using IQR method
numerical_features = [ 'price','area', 'bedrooms', 'bathrooms', 'stories', 'parking']
for feature in numerical_features:
    # Remove outliers using IQR method
    df[feature] = remove_outliers_iqr(df[feature])
````

````python
# Remove Stopwords
import pandas as pd
import nltk
from nltk.corpus import stopwords

# Download NLTK stopwords
nltk.download('stopwords')

# Load your dataset into DataFrame df
df = pd.read_csv('/content/Housing.csv')  # Replace 'your_dataset.csv' with the actual filename/path

# Function to remove stopwords from string columns
def remove_stopwords(data):
    stop_words = set(stopwords.words('english'))
    return data.apply(lambda x: x.apply(lambda y: ' '.join([word for word in y.split() if word.lower() not in stop_words])) if x.dtype == 'object' else x)

# Apply stopword removal to all string columns in the DataFrame
df_no_stopwords = remove_stopwords(df.copy())
````

#### 6. Rephrase Text

````python
# Rephrase Text
import nltk
nltk.download('punkt')
from nltk.corpus import wordnet
import random

# Download NLTK resources
nltk.download('wordnet')

# Function to rephrase text
def rephrase_text(text):
    tokens = nltk.word_tokenize(text)
    rephrased_tokens = []
    for token in tokens:
        synonyms = []
        for syn in wordnet.synsets(token):
            for lemma in syn.lemmas():
                synonyms.append(lemma.name())
        if synonyms:
            rephrased_tokens.append(random.choice(synonyms))
        else:
            rephrased_tokens.append(token)
    return ' '.join(rephrased_tokens)

# Original dataset description
dataset_description = "The dataset consists of several features related to houses, including: price, area, bedrooms, bathrooms, stories, mainroad, guestroom, basement, hotwaterheating, airconditioning, parking, prefarea, furnishingstatus."

# Rephrase the dataset description
rephrased_description = rephrase_text(dataset_description)

print("Original Description:")
print(dataset_description)
print("\nRephrased Description:")
print(rephrased_description)
````

## ***7. ML Model Implementation***

### ML Model - 1

````python
# ML Model - 1 Implementation

# Fit the Algorithm

# Predict on the model
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

data = pd.read_csv("Housing.csv")

categorical_features = ["mainroad", "guestroom", "basement", "hotwaterheating", "airconditioning", "prefarea", "furnishingstatus"]
numerical_features = ["area", "bedrooms", "bathrooms", "stories", "parking"]

numerical_transformer = StandardScaler()

categorical_transformer = OneHotEncoder()

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

model = KNeighborsRegressor(n_neighbors=5)  # You can adjust the number of neighbors as needed
my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),('model', model)])

y = data.price
X = data.drop('price', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=12)

my_pipeline.fit(X_train, y_train)

y_pred = my_pipeline.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("R-squared error:", r2)
print("Model saved successfully.")
````

### ML Model - 2

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

data = pd.read_csv("Housing.csv")

categorical_features = ["mainroad", "guestroom", "basement", "hotwaterheating", "airconditioning", "prefarea", "furnishingstatus"]
numerical_features = ["area", "bedrooms", "bathrooms", "stories", "parking"]

numerical_transformer = StandardScaler()

categorical_transformer = OneHotEncoder()

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

model = RandomForestRegressor(n_estimators=100, random_state=73)
my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),('model', model)])

y = data.price
X = data.drop('price', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=73)

my_pipeline.fit(X_train, y_train)

y_pred = my_pipeline.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("R-squared error:", r2)
print("Model saved successfully.")
````

#### 1. Explain the ML Model used and it's performance using Evaluation metric Score Chart.

````python
# Visualizing evaluation Metric Score chart
from sklearn.metrics import mean_absolute_error, median_absolute_error

# Plotting actual vs predicted prices
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual vs Predicted Price')
plt.show()

mae = mean_absolute_error(y_test, y_pred)
medae = median_absolute_error(y_test, y_pred)

# Print all evaluation metrics
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
print("Median Absolute Error (MedAE):", medae)
print("R-squared Error:", r2)

````

### ML Model - 3

````python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

data = pd.read_csv("Housing.csv")

categorical_features = ["mainroad", "guestroom", "basement", "hotwaterheating", "airconditioning", "prefarea", "furnishingstatus"]
numerical_features = ["area", "bedrooms", "bathrooms", "stories", "parking"]

numerical_transformer = StandardScaler()

categorical_transformer = OneHotEncoder()

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

model = LinearRegression()
my_pipeline = Pipeline(steps=[('preprocessor', preprocessor),('model', model)])

y = data.price
X = data.drop('price', axis=1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=73)

my_pipeline.fit(X_train, y_train)

y_pred = my_pipeline.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("R-squared error:", r2)
print("Model saved successfully.")

````
